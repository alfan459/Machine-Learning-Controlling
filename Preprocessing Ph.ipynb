```python
import os
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

df = pd.read_csv('phdown2.csv', delimiter=',')
df
```

```python
dataset = df[['Volume', 'Waktu', 'DeltapH']]
dataset
```

# Volume 40

```python
# Membuat DataFrame dari data
df = dataset[dataset["Volume"]==40]

# Definisikan fungsi eksponensial
def exp_func(x, a, b):
    return a * np.exp(b * x)

# Fit model eksponensial pada data
params, covariance = curve_fit(exp_func, df['Waktu'], df['DeltapH'], p0=(1, 0.1))

# Menghitung prediksi
df['DeltapH_Prediksi'] = exp_func(df['Waktu'], *params)

# Menghitung deviasi dari model
df['Deviasi'] = np.abs(df['DeltapH'] - df['DeltapH_Prediksi'])

# Filter data dengan deviasi kecil (misalnya < 0.1)
filtered_df = df[df['Deviasi'] < 0.1]

print("Data setelah memfilter berdasarkan pola eksponensial:")
filtered_df
average_df = filtered_df.groupby(['Volume', 'Waktu']).mean()[['DeltapH_Prediksi']]
df_40 = average_df.reset_index()
df_40
```
