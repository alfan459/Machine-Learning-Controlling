```python
import os
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

data = pd.read_csv('phdown2.csv', delimiter=',')
data1= data.groupby(['Volume', 'Waktu']).mean()[['DeltapH']]
df = data1.reset_index()
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot scatter DeltapH dengan volume sebagai warna
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['Waktu'], df['DeltapH'], c=df['Volume'], cmap='viridis', s=100, alpha=0.8)
plt.colorbar(scatter, label='Volume')
plt.title('Scatter Plot DeltapH terhadap Waktu Dengan data Eskponensial')
plt.xlabel('Waktu')
plt.ylabel('DeltapH')
plt.grid(True)
plt.show()
```

# Volume 40

```python
# Membuat DataFrame dari data
data40 = df[df["Volume"]==40]

# Definisikan fungsi eksponensial
def exp_func(x, a, b):
    return a * np.exp(b * x)

# Fit model eksponensial pada data
params, covariance = curve_fit(exp_func, data40['Waktu'], data40['DeltapH'], p0=(1, 0.1))

# Menghitung prediksi
data40['DeltapH_Prediksi'] = exp_func(data40['Waktu'], *params)

# Menghitung deviasi dari model
data40['Deviasi'] = np.abs(data40['DeltapH'] - data40['DeltapH_Prediksi'])

# Filter data dengan deviasi kecil (misalnya < 0.1)
filtered_df = data40[data40['Deviasi'] < 0.1]

print("Data setelah memfilter berdasarkan pola eksponensial:")
filtered_df
average_df = filtered_df.groupby(['Volume', 'Waktu']).mean()[['DeltapH_Prediksi']]
df_40 = average_df.reset_index()
df_40
```
