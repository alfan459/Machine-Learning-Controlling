```python
import os
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

df = pd.read_csv('nutA1.csv', delimiter=';')
df
```

```python
dataset = df[["Volume Tandon (L)", "Waktu (s)", "Delta TDS (ppm)"]]
dataset
```

# Volume 40

```python
import pandas as pd

# Membuat DataFrame dari data
df = dataset[dataset["Volume Tandon (L)"]==40]

# Menentukan alpha untuk smoothing eksponensial
alpha = 0.3

# Mengaplikasikan smoothing eksponensial pada DeltapH
df['DeltapH_Smooth'] = df["Delta TDS (ppm)"].ewm(alpha=alpha, adjust=False).mean()

print("Data setelah menerapkan smoothing eksponensial:")
df
average_df = df.groupby(["Volume Tandon (L)", "Waktu (s)"]).mean()[["DeltapH_Smooth"]]
df_40 = average_df.reset_index()
df_40
```
